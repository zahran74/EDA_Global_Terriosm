```python
# Load the dataset into a Dask DataFrame


dask_terrorism_df = dd.read_csv(r'F:\ITI\Python libraries\EDA_Global_Terriosm\terrorism_cleaned.csv')

# Perform similar operations as above with Dask
ddf_attacks_by_year = dask_terrorism_df.groupby('year')['nkill', 'nwound'].sum().compute()
ddf_attacks_by_region = dask_terrorism_df.groupby('region')['nkill', 'nwound'].sum().compute()
ddf_attacks_by_country = dask_terrorism_df.groupby('country')['nkill', 'nwound'].sum().compute()
ddf_attacks_by_attack_type = dask_terrorism_df.groupby('attacktype')['nkill', 'nwound'].sum().compute()
ddf_attacks_by_target_type = dask_terrorism_df.groupby('targtype')['nkill', 'nwound'].sum().compute()

print("Dask DataFrame Results:")
print(ddf_attacks_by_year)
print(ddf_attacks_by_region)
print(ddf_attacks_by_country)
print(ddf_attacks_by_attack_type)
print(ddf_attacks_by_target_type)
```

```python
dask_terrorism_df
```

```python
# Memory usage comparison

# Pandas memory usage
pandas_memory = terrorism_df.memory_usage(deep=True).sum()
print(f"Pandas memory usage: {pandas_memory / 1024**2} MB")

# Dask memory usage
dask_memory = dask_terrorism_df.memory_usage().compute().sum()
print(f"Dask memory usage: {dask_memory / 1024**2} MB")
```

```python
# Time Comparison

# Pandas
start_time = time.time()
terrorism_df_filtered = terrorism_df[terrorism_df['country'] == 'Afghanistan']
pandas_time = time.time() - start_time
print(f"Pandas filtering time: {pandas_time:} seconds")

# Dask
start_time = time.time()
dask_terrorism_df_filtered = dask_terrorism_df[dask_terrorism_df['country'] == 'Afghanistan']
dask_time = time.time() - start_time
print(f"Dask filtering time: {dask_time:} seconds")
```

```python
# Line plot showing the trend of terrorist attacks over the years
plt.figure(figsize=(15, 12))
attacks_per_year.plot(kind = 'line', color= 'red')
plt.title('Trend of Terrorist Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Number Of Attacks')
plt.show()
```

```python
# Bar plot of the number of attacks by region
plt.figure(figsize = (12, 6))
attacks_per_region.plot(kind= 'bar', color= 'green')
plt.title('Number of Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Number Of Attacks')
plt.show()
```

```python
# Heatmap to visualize the correlation between different features
plt.figure(figsize=(10, 6))
sns.heatmap(terrorism_df[['nkill', 'nwound']].corr(), annot=True, cmap='YlOrRd')
plt.title('Correlation Heatmap Between Different Features')
plt.show()
```

```python
# Scatter plot showing the relationship between the number of casualties and the type of attack
plt.figure(figsize=(12, 8))
sns.scatterplot(x='attacktype', y='nkill', data=terrorism_df)
plt.title('Relationship Between Attack Type and Number of Casualties')
plt.xlabel('Attack Type')
plt.ylabel('Number of Casualties')
plt.show()
```

```python
# Most Attcaked Target Types #

plt.figure(figsize=(20, 8))
sns.set(font_scale=1.0)

top_10_targets = terrorism_df['targtype'].value_counts().nlargest(10).index

xaxis = sns.countplot(x='targtype', data=terrorism_df, order=top_10_targets)
xaxis.set_xticklabels(xaxis.get_xticklabels(), rotation=60)

plt.xlabel('Target Types', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Top 10 Most Target Attacked Types', fontsize=16)
plt.show()
```

```python
# Most Gangs make Attacks #


plt.figure(figsize=(20, 8))
sns.set(font_scale=1.0)

top_10_Gangs = terrorism_df['gname'].value_counts().nlargest(10).index

xaxis = sns.countplot(x='gname', data=terrorism_df, order=top_10_Gangs)
xaxis.set_xticklabels(xaxis.get_xticklabels(), rotation=60)

plt.xlabel('Gang Names', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Top 10 Most Gangs make Attacks', fontsize=16)
plt.show()
```

```python
#Top 10 Contries with most number of Attacks by terror groups#

fig= plt.figure(figsize=(20, 10))
sns.set(font_scale=0.9)
terror_country = sns.barplot(x=terrorism_df['country'].value_counts()[0:50].index, y=terrorism_df['country'].value_counts()[0:50], palette='RdYlGn')
terror_country.set_xticklabels(terror_country.get_xticklabels(), rotation=70)
terror_country.set_xlabel('Country', fontsize=20)
terror_country.set_ylabel('Counts', fontsize=20)
plt.title('Top 10 Countries: Most Attacks by Terrorist Groups', fontsize=20)
```

```python
# Most cities Attacked # 

plt.figure(figsize=(20, 8))
sns.set(font_scale=1.0)

most_cities = terrorism_df['city'].value_counts().nlargest(10).index

xaxis = sns.countplot(x='city', data=terrorism_df, order=most_cities)
xaxis.set_xticklabels(xaxis.get_xticklabels(), rotation=60)

plt.xlabel('City', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Most cities Attacked', fontsize=16)
plt.show()
```
